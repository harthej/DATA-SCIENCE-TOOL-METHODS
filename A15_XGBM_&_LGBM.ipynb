# ***XGBM***

```python

X = train_data.iloc[:, [2, 4, 5, 6, 7]] #Selecting the desired columns


# Access the 'Survived' column of the 'train_data' DataFrame using iloc
Y = train_data.iloc[:, 1] # Selecting the 2nd column ('Survived')


# split data into train and test sets
seed = 7
test_size = 0.33
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=test_size, random_state=seed)

# Convert 'Sex' column to numerical using one-hot encoding
X_train = pd.get_dummies(X_train, columns=['Sex'], drop_first=True) # drop_first to avoid multicollinearity
X_test = pd.get_dummies(X_test, columns=['Sex'], drop_first=True)

# fit model no training data
model = XGBClassifier()
model.fit(X_train, y_train)
```

```python
# make predictions for test data
y_pred = model.predict(X_test)
predictions = [round(value) for value in y_pred]
```

```python
accuracy = accuracy_score(y_test, predictions)
print("Accuracy: %.2f%%" % (accuracy * 100.0))
```

```python
import pandas as pd
import lightgbm as lgb
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder


# Create a LabelEncoder object
label_encoder = LabelEncoder()

# Fit the encoder to the 'Sex' column and transform it
x_train['Sex'] = label_encoder.fit_transform(x_train['Sex'])
x_test['Sex'] = label_encoder.transform(x_test['Sex'])  # Use transform on test data


# Now you can proceed with creating the LightGBM dataset and training the model
d_train = lgb.Dataset(x_train, label=y_train)
```
