```python
# Define the column names for the mushroom dataset
names = ['class', 'cap-shape', 'cap-surface', 'cap-color', 'bruises', 'odor', 'gill-attachment', 'gill-spacing', 'gill-size', 'gill-color', 'stalk-shape', 'stalk-root', 'stalk-surface-above-ring', 'stalk-surface-below-ring', 'stalk-color-above-ring', 'stalk-color-below-ring', 'veil-type', 'veil-color', 'ring-number', 'ring-type', 'spore-print-color', 'population', 'habitat']

# Assuming 'mushroom.csv' is in the current directory
filename = 'mushroom.csv'

# Now read the CSV file with the specified column names
mushroom = pd.read_csv(filename, names=names)  # Using pd.read_csv since you already imported pandas as pd

# Create the feature matrix (X) and target vector (Y)
X = mushroom.drop('class', axis=1)  # Assuming 'class' is the target variable
Y = mushroom['class']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,Y, test_size = 0.3)
```

```python

X_train.shape, y_train.shape, X_test.shape, y_test.shape
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.svm import SVC
from sklearn.preprocessing import LabelEncoder

# Create a LabelEncoder object
encoder = LabelEncoder()

# Iterate through all columns in X_train and encode categorical features
for col in X_train.columns:
    if X_train[col].dtype == 'object':  # Check if the column is of object type (likely string)
        # Fit the encoder on the combined unique values from both train and test sets
        all_values = pd.concat([X_train[col], X_test[col]]).unique()
        encoder.fit(all_values)

        X_train[col] = encoder.transform(X_train[col])
        X_test[col] = encoder.transform(X_test[col])  # Apply the same encoding to X_test

clf = SVC()
param_grid = [{'kernel':['rbf'],'gamma':[50,5,10,0.5],'C':[15,14,13,12,11,10,0.1,0.001] }]
gsv = GridSearchCV(clf,param_grid,cv=10)
gsv.fit(X_train,y_train)
```

```python
gsv.best_params_ , gsv.best_score_
```

```python
clf = SVC(C= 15, gamma = 50)
clf.fit(X_train , y_train)
y_pred = clf.predict(X_test)
acc = accuracy_score(y_test, y_pred) * 100
print("Accuracy =", acc)
confusion_matrix(y_test, y_pred)
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.svm import SVC
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import accuracy_score


# Define the kernels to compare
kernels = ['linear', 'poly', 'rbf']

# Store results for comparison
results = []

for kernel in kernels:
    # Create and train the SVM model
    clf = SVC(kernel=kernel)
    clf.fit(X_train, y_train)

    # Make predictions and evaluate
    y_pred = clf.predict(X_test)
    accuracy = accuracy_score(y_test, y_pred)

    # Store the results
    results.append({'kernel': kernel, 'accuracy': accuracy})

# Display the results
results_df = pd.DataFrame(results)
print(results_df)
```
